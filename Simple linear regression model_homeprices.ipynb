```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('F:/datasets/homeprices.csv')
```

```python
data
```

```python
plt.xlabel('Area')
plt.ylabel("Price")
plt.scatter(data.area, data.price, color='blue', marker = '*')
```

```python
x = data.drop('price', axis=1) # feature
y = data['price'] # target
```

```python
x
```

```python
y
```

```python
model = LinearRegression() # creating object of the Linear Regression class
model.fit(x,y) # model training
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(x)
y_pred
```

```python
pd.DataFrame({'True y values':y, 'Predicted y values':y_pred})
```

```python
plt.scatter(x,y)
plt.xlabel('Area')
plt.ylabel('Price')
plt.plot(data['area'], y_pred, color = 'red')
```

```python
model.predict([[3700]])
```

```python
model.predict([[2600],[2700],[5000]])
```

```python
model.score(x,y)
```
